### Student Grade Calculator

```python
sub1 = float(input('Enter marks of subject 1: '))
sub2 = float(input('Enter marks of subject 2: '))
sub3 = float(input('Enter marks of subject 3: '))
sub4 = float(input('Enter marks of subject 4: '))
sub5 = float(input('Enter marks of subject 5: '))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
print(f'\nTotal Marks of 5 Subjects: {total_marks}')

marks = total_marks / 5
print(f'\nAverage Marks: {marks}')
print('-'*20)

if marks >= 80 and marks <= 100:
    print('Grade A')
elif marks >= 60 and marks < 80:
    print('Grade B')
elif marks >= 50 and marks < 60:
    print('Grade C')
elif marks >= 40 and marks < 50:
    print('Grade D')
elif marks >= 0 and marks < 40:
    print('Failed')
else:
    print('Invalid Marks. Marks must be in range of 0 to 100.')
```

## ATM Amount Withdrawl

```python
amount = 50000
user_pin = 1234

pin = int(input('Enter your pin number: '))
balance = int(input('Enter balance amount for withdrawl: '))

if pin == user_pin and balance < amount:
    print(f'Rs.{balance} withdrawn from A/C 12345678.')
    remaining_balance = amount - balance
    print(f'Remaining Balance: Rs.{remaining_balance}')
elif pin == user_pin and balance > amount:
    print(f'Insufficient Balance.')
elif pin != user_pin and balance < amount:
    print(f'Incorrect Pin.')
else:
    print('Incorrect Pin Number or Insufficient Balance Amount.')
```

```python
'''
calculate tax based on annual income 
<=250000 then no tax 
250001 - 500000 then 5% tax 
500001 - 1000000 then 20% 
more than 1000000 then 30% 

output
-----------
Annual Income: Rs.100000
Tax Payable: Rs.5000
Remaining Salary: Rs.95000
'''
```

```python
annual_income = int(input('Enter annual income: '))

if annual_income > 0 and annual_income <= 250000:
    print(f'Annual Income: Rs.{annual_income}')
    print(f'Tax Payable (0%): Rs.0')
    print(f'Net Salary: {annual_income}')
elif annual_income > 250000 and annual_income <= 500000:   
    print(f'Annual Income: Rs.{annual_income}')
    taxable = annual_income * 0.05
    print(f'Tax payable (5%): Rs.{taxable}')
    net_sal = annual_income - taxable
    print(f'Net Salary: Rs.{net_sal}')
elif annual_income > 500000 and annual_income <= 1000000:
    print(f'Annual Income: Rs.{annual_income}')
    taxable = annual_income * 0.20
    print(f'Tax payable (20%): Rs.{taxable}')
    net_sal = annual_income - taxable
    print(f'Net Salary: Rs.{net_sal}')
elif annual_income > 1000000:
    print(f'Annual Income: Rs.{annual_income}')
    taxable = annual_income * 0.30
    print(f'Tax payable (30%): Rs.{taxable}')
    net_sal = annual_income - taxable
    print(f'Net Salary: Rs.{net_sal}')
else:
    print('Amount must be more than 0 or Positive Integer.')
```

## Nested If

```python
# A student is eligible for an exam if: 
# Attendance >= 75 
# internal marks >= 40 
# print exact reason if not eligible
```

```python
attendance = int(input('Enter attendance value: '))
internal_marks = float(input('Enter internal marks: '))

if attendance >= 75 and internal_marks >= 40:
    print('You are eligible for examination.')
elif attendance < 75 and internal_marks >= 40:
    print('You are not eligible for examination. Attendance requirement does not meet.')
elif attendance >= 75 and internal_marks < 40:
    print('You are not eligible for examination. Internal marks requirement does not meet.')
else:
    print('You are not eligible for examination.')
```

```python
attendance = int(input('Enter attendance value: '))

if attendance >= 75:
    internal_marks = float(input('Enter internal marks: '))
    if internal_marks >= 40:
        print('You are eligible for examination.')
    else:
        print('You are not eligible for examination. Internal marks requirement does not meet.')
else:
    print('You are not eligible for examination. Attendance requirement does not meet.')
```

```python
# Switch Case
# --> Match Case
# 1 - 7 -> Sunday to Saturday
```

```python
number = int(input('Enter a number (1 - 7): '))

if number == 1:
    print('Sunday')
elif number == 2:
    print('Monday')
elif number == 3:
    print('Tuesday')
elif number == 7:
    print('Saturday')
else:
    print('Invalid Number..')
```

```python
# match case
number = input('Enter a number: ')

match number:
    case '1':  # 4 == 1
        print('Sunday')
    case '2': # 4 == 2
        print('Monday')
    case '3': # 4 == 3
        print('Tuesday')
    case '4': # 4 == 4
        print('Wednesday')
    case '5':
        print('Thursday')
    case '6':
        print('Friday')
    case '7':
        print('Saturday')
    case _: # default case
        print('Invalid Number.')
```
