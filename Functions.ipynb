```python
def annualIncome(annual_income):
    if annual_income > 0 and annual_income <= 250000:
        print(f'Annual Income: Rs.{annual_income}')
        print(f'Tax Payable (0%): Rs.0')
        print(f'Net Salary: {annual_income}')
    elif annual_income > 250000 and annual_income <= 500000:   
        print(f'Annual Income: Rs.{annual_income}')
        taxable = annual_income * 0.05
        print(f'Tax payable (5%): Rs.{taxable}')
        net_sal = annual_income - taxable
        print(f'Net Salary: Rs.{net_sal}')
    elif annual_income > 500000 and annual_income <= 1000000:
        print(f'Annual Income: Rs.{annual_income}')
        taxable = annual_income * 0.20
        print(f'Tax payable (20%): Rs.{taxable}')
        net_sal = annual_income - taxable
        print(f'Net Salary: Rs.{net_sal}')
    elif annual_income > 1000000:
        print(f'Annual Income: Rs.{annual_income}')
        taxable = annual_income * 0.30
        print(f'Tax payable (30%): Rs.{taxable}')
        net_sal = annual_income - taxable
        print(f'Net Salary: Rs.{net_sal}')
    else:
        print('Amount must be more than 0 or Positive Integer.')
```

```python
annualIncome(300000)
```

### All Arguments (*args) and All Keyword Arguments (**kwargs)

```python
def details(*details):
    print(details)
    for a in details:
        print(a)
```

```python
details('Ram', 30, 'Kathamandu', 'Masters', 1234567890)
```

```python
def kw_details(**empDetails):
    for key, val in empDetails.items():
        print(f'{key} of Employees: {val}')
```

```python
kw_details(Name='Ram', Age=20, Qualification='Bachelors', Phone=1234567890)
```

### Nested Function

```python
def add(x, y):
    print(f'Addition of {x} and {y} is: {x+y}')

def sub(x, y):
    print(f'Subtraction of {x} and {y} is: {x-y}')

def mul(x, y):
    print(f'Multiplication of {x} and {y} is: {x*y}')

def div(x, y):
    print(f'Division of {x} and {y} is: {x/y}')
```

```python
def calculate():
    x = int(input('Enter first number: '))
    y = int(input('Enter second number: '))
    add(x, y)
    sub(x, y)
    mul(x, y)
    div(x, y)
```

```python
calculate()
```

```python
def add(x, y):
    return f'Addition of {x} and {y} is: {x+y}'
    
def sub(x, y):
    return f'Subtraction of {x} and {y} is: {x-y}'

def mul(x, y):
    return f'Multiplication of {x} and {y} is: {x*y}'

def div(x, y):
    return f'Division of {x} and {y} is: {x/y}'
```

```python
def calculator(x, y):
    add(x, y)
    sub(x, y)
    mul(x, y)
    div(x, y)
    return add(x, y), sub(x, y), mul(x, y), div(x, y)
```

```python
calculator(5, 6)
```

### Lambda Function

```python
# function_name = lambda arg1, arg2 : calculation... / condition...
```

```python
add_num = lambda x, y, z: x + y + z
add_num(5, 10, 15)
```

```python
odd_even = lambda num : "Even" if num % 2 == 0 else "Odd"
```

```python
print(odd_even(7))
print(odd_even(8))
```

```python
def odd_even(num):
    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
```
