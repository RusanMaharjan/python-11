### Constructor

```python
class Student:
    def __init__(self, name, section, no_of_subjects, past_faliures, current_grade):
        self.name = name
        self.section = section
        self.no_of_subjects = no_of_subjects
        self.past_failures = past_faliures
        self.current_grade = current_grade
        self.school_name = 'ABC School'
        self.school_location = 'Kumaripati, Lalitpur'

    def student_details(self):
        print(f'School Name: {self.school_name}')
        print(f'School Location: {self.school_location}')
        print(f'Name of Student: {self.name}')
        print(f'Section of {self.name}: {self.section}')
        print(f'Number of Subjects: {self.no_of_subjects}')
        print(f'Past Faliures: {self.past_failures}')
        print(f'Current Grade of {self.name}: {self.current_grade}')
```

```python
s = Student('Bob', 'C11', 4, 2, 'A')
```

```python
s.student_details()
```

```python
class SmartCalculator:
    def __init__(self, num1, num2, num3):
        self.num1 = num1
        self.num2 = num2
        self.num3 = num3

    def add(self):
        return f'Sum of {self.num1} and {self.num3} is: {self.num1 + self.num3}'

    def is_even(self):
        if self.num2 % 2 == 0:
            return 'Even'
        else:
            return 'Odd'
```

```python
sc = SmartCalculator(20, 31, 6)
```

```python
sc.add()
```

```python
sc.is_even()
```

## Inheritance

```python
# Multiple Classes
'''
Types of Inheritance
-----------------------
1. Simple Inheritance
2. Multiple Inheritance
3. Multi-Level Inheritance
'''
'''
Parent and Child
Child Class Inherit / extends Parent Class

one school has many students
'''
```

```python
# parent class
class School:
    def __init__(self, school_name, school_location, school_contact):
        self.school_name = school_name
        self.school_location = school_location
        self.school_contact = school_contact

    def school_details(self):
        print(f'Name of school: {self.school_name}')
        print(f'Location of School: {self.school_location}')
        print(f'School Contact: {self.school_contact}')
```

```python
# child class
class Student(School): # Class Student inherits / extends School class
    def __init__(self, student_name, student_section, student_past_failures, student_contact, student_current_grade,
                school_name, school_location, school_contact):
        # super() -> super keyword is used to inherit parent class constructor
        super().__init__(school_name, school_location, school_contact)
        self.student_name = student_name
        self.student_section = student_section
        self.student_past_failures = student_past_failures
        self.student_contact = student_contact
        self.student_current_grade = student_current_grade

    def student_details(self):
        print(f'Name of student: {self.student_name}')
        print(f'Section of Student: {self.student_section}')
        print(f'Students Past Failures: {self.student_past_failures}')
        print(f'Student Contact Number: {self.student_contact}')
        print(f'Student Current Grade: {self.student_current_grade}')
```

```python
s = Student('Charles', 'C7', 2, 98765433220, 'B+', 'ABC School', 'Kathmandu', '01-5250176')
```

```python
s.student_details()
s.school_details()
```

```python
# Child class
class Teacher(School):
    def __init__(self, teacher_name, teacher_contact, total_classes, teacher_qualification, 
                 school_name, school_location, school_contact):
        super().__init__(school_name, school_location, school_contact)
        self.teacher_name = teacher_name
        self.teacher_contact = teacher_contact
        self.total_classes = total_classes
        self.teacher_qualification = teacher_qualification

    def teacher_details(self):
        print(f'Teacher Name: {self.teacher_name}')
        print(f'Teacher Contact: {self.teacher_contact}')
        print(f'Total Classes: {self.total_classes}')
        print(f'Teacher Qualification: {self.teacher_qualification}')
```

```python
t = Teacher('Bob', '1234567890', 4, 'Bachelors', 'ABC School', 'Kathmandu', '01-5250176')
t.teacher_details()
t.school_details()
```

## Multi-Level Inheritance

```python
# Student -> teacher -> School
```

```python
# parent class
class School:
    def __init__(self, school_name, school_location, school_contact):
        self.school_name = school_name
        self.school_location = school_location
        self.school_contact = school_contact

    def school_details(self):
        print(f'Name of school: {self.school_name}')
        print(f'Location of School: {self.school_location}')
        print(f'School Contact: {self.school_contact}')
```

```python
class Teacher(School):
    def __init__(self, school_name, school_location, school_contact, teacher_name, teacher_address, 
                 teacher_contact, total_classes, teacher_qualification):
        super().__init__(school_name, school_location, school_contact)
        self.teacher_name = teacher_name
        self.teacher_address = teacher_address
        self.teacher_contact = teacher_contact
        self.total_classes = total_classes
        self.teacher_qualification = teacher_qualification

    def teacher_details(self):
        print(f'Teacher Name: {self.teacher_name}')
        print(f'Teacher Address: {self.teacher_address}')
        print(f'Teacher Contact: {self.teacher_contact}')
        print(f'Total Classes: {self.total_classes}')
        print(f'Teacher Qualification: {self.teacher_qualification}')
```

```python
class Student(Teacher):
    def __init__(self, student_name, student_section, student_past_failures, student_contact, student_current_grade,
                school_name, school_location, school_contact,
                 teacher_name, teacher_address, teacher_contact, total_classes, teacher_qualification):
        super().__init__(school_name, school_location, school_contact, teacher_name, teacher_address, 
                 teacher_contact, total_classes, teacher_qualification)
        self.student_name = student_name
        self.student_section = student_section
        self.student_past_failures = student_past_failures
        self.student_contact = student_contact
        self.student_current_grade = student_current_grade

    def student_details(self):
        print(f'Name of student: {self.student_name}')
        print(f'Section of Student: {self.student_section}')
        print(f'Students Past Failures: {self.student_past_failures}')
        print(f'Student Contact Number: {self.student_contact}')
        print(f'Student Current Grade: {self.student_current_grade}')

        print(f'Name of school: {self.school_name}')
        print(f'Location of School: {self.school_location}')
        print(f'School Contact: {self.school_contact}')

        print(f'Teacher Name: {self.teacher_name}')
        print(f'Teacher Address: {self.teacher_address}')
        print(f'Teacher Contact: {self.teacher_contact}')
        print(f'Total Classes: {self.total_classes}')
        print(f'Teacher Qualification: {self.teacher_qualification}')
```

```python
s = Student('John', 'C11', 0, 1234567890, 'B+', 'ABC School', 'Kathamndu', '01-1234567', 'Ram', 'Lalitpur', 
            9876543210, 3, 'Bachelors')
```

```python
s.student_details()
```

```python
s.student_name
```

## Encapsulation

```python
'''
Data Hiding
-------------
Access Modifiers
1. Private
2. Public
3. Protected

self.name -> Public
__self.name -> private
_self.name -> protected
'''
```

```python
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone, customer_email):
        self.__customer_id = customer_id
        self.__customer_name = customer_name
        self.__customer_address = customer_address
        self.__customer_phone = customer_phone
        self.__customer_email = customer_email

    def customer_details(self):
        print(f'Customer ID: {self.__customer_id}')
        print(f'Customer Name: {self.__customer_name}')
        print(f'Customer Address: {self.__customer_address}')
        print(f'Customer Phone: {self.__customer_phone}')
        print(f'Customer Email: {self.__customer_email}')
```

```python
class Orders(Customer):
    def __init__(self, order_id, order_date, order_time, order_destination,
                customer_id, customer_name, customer_address, customer_phone, customer_email):
        super().__init__(customer_id, customer_name, customer_address, customer_phone, customer_email)
        self.__order_id = order_id
        self.__order_date = order_date
        self.__order_time = order_time
        self.__order_destination = order_destination

    def order_details(self):
        print(f'Order ID: {self.__order_id}')
        print(f'Order Date: {self.__order_date}')
        print(f'Order Time: {self.__order_time}')
        print(f'Order Destination: {self.__order_destination}')
```

```python
o = Orders(1, '20th May, 2026', '12:30 PM', 'Bhaktapur', 12, 'Ram', 'Lalitpur', 1234567890, 'ram@gmail.com')
```

```python
o.order_details()
o.customer_details()
```

```python
o.__order_id
```

### Protected

```python
class Brand:
    def __init__(self, brand_id, brand_name):
        self._brand_id = brand_id
        self._brand_name = brand_name

    def brand_details(self):
        print(f'Brand ID: {self._brand_id}')
        print(f'Brand Name: {self._brand_name}')
```

```python
class Product(Brand):
    def __init__(self, product_id, product_name, product_desc, product_price, brand_id, brand_name):
        super().__init__(brand_id, brand_name)
        self._product_id = product_id
        self._product_name = product_name
        self._product_desc = product_desc
        self._product_price = product_price

    def product_details(self):
        print(f'Product ID: {self._product_id}')
        print(f'Product Name: {self._product_name}')
        print(f'Product Description: {self._product_desc}')
        print(f'Product Price: Rs.{self._product_price}')
```

```python
p = Product(1,'Nokia Mobile', 'Keypad Mobile', 3000, 123, 'Nokia')
```

```python
p.product_details()
p.brand_details()
```

```python
p._brand_id
```

## Polymorphism

### Message Gateway

```python
class Facebook:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'Facebook:\n{send_by} sent you a message.')
        print(f'Message: {message}')

class WhatsApp:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'WhatsApp:\n{send_by} sent you a message.')
        print(f'Message: {message}')

class Viber:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'Viber:\n{send_by} sent you a message.')
        print(f'Message: {message}')
```

```python
def message(message_object, send_by, send_to, message):
    message_object.send_message(send_by, send_to, message)
```

```python
message(WhatsApp(), 'Ram', 'Shyam', 'Hello!!')
```

```python
obj = [Facebook(), Viber(), WhatsApp()]
for o in obj:
    message(o, 'Ram', 'Shyam', 'Hello!!')
    print('*'*15)
```

## Abstraction

```python
'''
Abstraction
Interface -> A class without body.

- If we create abstract class then its atleast one method/function must be abstract function.
- Abstract function must be empty.
'''
```

```python
# Abstract Base Class
from abc import ABC, abstractmethod
```

```python
class MobileBanking(ABC):
    @abstractmethod
    def statement(self):
        pass
        
    @abstractmethod
    def user_account_details(self):
        pass

    def connection(self):
        return 'Mobile Banking Server Connected to Database.'

    def username(self, uname):
        return f'{uname}'
```

```python
class ShowDetails(MobileBanking):
    def show_statement(self):
        return f'Choose Statement'

    def statement(self):
        return 'Statement From Date1 to Date2.'

    def user_account_details(self):
        return 'User Account'
```

```python
s = ShowDetails()
```

```python
s.statement()
```

```python
print(s.connection())
print(s.username('Bob'))
```
