```python
'''
Operations
------------
r -> Read Operation
w -> Write Operation
a -> Append Operation
rb -> Read Binary Operation -> copy
wb -> Write Binary Operation -> paste
'''
```

## Text File

```python
# open(file_name, operation)
contents = input('Enter your content here: ')
file = open('text_file1.txt', 'w')
file.write(contents)
file.close()
```

```python
with open('text_file_new.txt', 'w') as file:
    file.write('This is new text file.')
```

### Read Text Data

```python
with open('text_file1.txt', 'r') as file:
    content = file.read()
    print(content)
```

### Append Text

```python
contents = input('Enter your content here: ')
with open('text_file1.txt', 'a') as file:
    file.write(contents)
```

## CSV File

```python
import csv
```

```python
countries = [
    ['Name', 'Destination', 'Price'],
    ['USA', 'New York', '$10,000'],
    ['UK', 'London', '$12,000'],
    ['Italy', 'Rome', '$9,000'],
    ['Nepal', 'Chitwan', '$3,000']
]
```

```python
## CSV/Dictionary -> open(file_name, operation, newline="")
with open('countries.csv', 'w', newline="") as file:
    writer = csv.writer(file) # Create file
    writer.writerows(countries) # insert data into file
```

```python
with open('countries.csv', 'r') as file:
    reader = csv.reader(file) # Read File
    for row in reader: # Loop data and print data
        print(row)
```

```python
new_countries = [
    ['India', 'New Delhi', '$4,000'],
    ['Nepal', 'Pokhara', '$4,000']
]
with open('countries.csv', 'a', newline="") as file:
    writer = csv.writer(file)
    writer.writerows(new_countries)
```

```python
personal_details = [
    {
        'Name': 'Ram',
        'Age': 30,
        'Position': 'Data Scientist'
    },
    {
        'Name': 'Bob',
        'Age': 40,
        'Position': 'Data Analyst'
    },
    {
        'Name': 'Edward',
        'Age': 35,
        'Position': 'Database Administrator'
    },
    {
        'Name': 'Charles',
        'Age': 20,
        'Position': 'Data Engineer Intern'
    }
]
```

```python
with open('personal_details.csv', 'w', newline='') as file:
    columns = ['Name', 'Age', 'Position'] # column names
    writer = csv.DictWriter(file, fieldnames=columns) # create file and set header content
    writer.writeheader() # insert header
    writer.writerows(personal_details) # insert data
```

```python
with open('personal_details.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        for key, val in row.items():
            print(f'{key}: {val}')
        print('-'*15)
```

```python
new_personal_details = [
    {
        'Name': 'John',
        'Age': 30,
        'Position': 'Back-End Developer'
    },
    {
        'Name': 'Bill',
        'Age': 25,
        'Position': 'Front-End Developer'
    }
]
```

```python
with open('personal_details.csv', 'a', newline='') as file:
    columns = ['Name', 'Age', 'Position']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writerows(new_personal_details)
```

## JSON (Javascript Object Notation) File

```python
import json
```

```python
personal_details_json = [
    {
        'Name': 'Ram',
        'Age': 30,
        'Position': 'Data Scientist'
    },
    {
        'Name': 'Bob',
        'Age': 40,
        'Position': 'Data Analyst'
    },
    {
        'Name': 'Edward',
        'Age': 35,
        'Position': 'Database Administrator'
    },
    {
        'Name': 'Charles',
        'Age': 20,
        'Position': 'Data Engineer Intern'
    }
]
```

```python
# file create -> dump
# file read -> load
with open("pd_info.json", 'w') as file:
    json.dump(personal_details_json, file, indent=4)
```

```python
with open("pd_info.json", 'r') as file:
    content = json.load(file)
    # print(content)
    for data in content:
        print(data)
```
