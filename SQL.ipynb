```python
import sqlite3
```

```python
conn = sqlite3.connect('BikeStores.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('''
    select * from customers;
''')
data = cursor.fetchall()
for row in data:
    print(row)
```

## Create Table

```python
try:
    cursor.execute('''
        Create table customers (
        	customer_id integer primary key autoincrement,
        	first_name text not null,
        	last_name text not null,
        	phone text not null,
        	email text not null,
        	street text,
        	city text not null
        );
    ''')
    print('Customer table created.')
except:
    print(f'Table Customer already exists.')
```

```python
try:
    cursor.execute('''
        Create table staff (
            staff_id integer primary key autoincrement,
            first_name text not null,
            last_name text not null,
            phone text not null,
            email text not null
        )
    ''')
    print('Staff table created.')
except:
    print(f'Table Staff already exists.')
```

```python
try:
    cursor.execute('''
        Create table orders (
        	order_id integer primary key autoincrement,
        	order_status text not null,
        	order_date date,
        	customer_id integer,
        	staff_id integer,
        	foreign key (customer_id) references customers(customer_id)
        	foreign key (staff_id) references staff(staff_id)
        );
    ''')
    print('Orders table created!')
except:
    print('Orders table already exists.')
```

## Insert Values

```python
try:
    cursor.execute('''
        Insert into customers (first_name, last_name, phone, email, street, city)
        values 
        ('Bob','Miles','87656789','milesBob@gmail.com', 'LA', 'California'),
        ('Charles','Morgan','345678765','CMorgan@gmail.com', 'Westminister', 'London');
    ''')
    conn.commit()
    print('Data inserted successfully.')
except NameError as e:
    print(e)
except:
    print('Data insertion failed.')
```

```python
try:
    cursor.execute('''
        Insert into staff (first_name, last_name, phone, email)
        values
        ('Sita', 'Kumari', '1234567', 'sita@gmail.com'),
        ('Jay', 'Kisan', '34567890', 'Jk23@gmail.com'),
        ('Shivam', 'Kumar', '456789876', 'shivam123@gmail.com')
    ''')
    conn.commit()
    print('Data inserted successfully.')
except:
    print('Data insertion failed.')
```

```python
try:
    cursor.execute('''
        Insert into orders (order_status, order_date, customer_id, staff_id)
        values
        ('Pending', '2026-05-10', 2, 1),
        ('Invalid', '2026-04-2', 1, 2),
        ('Invalid', '2026-05-27', 1, 3)
    ''')
    conn.commit()
    print('Data inserted successfully.')
except:
    print('Data insertion failed.')
```

## Update Table

```python
try:
    cursor.execute('''
        Update customers set first_name = 'RAAAMMMM', last_name = 'Dude' where customer_id = 1;
    ''')
    conn.commit()
    print('Data updated successfully.')
except:
    print('Data update failed.')
```

## Delete data

```python
try:
    cursor.execute('''
        delete from orders where order_id = 8;
    ''')
    conn.commit()
    print('Data deleted successfully.')
except:
    print('Data deletion failed.')
```
