```python
try:
    a = int(input('Enter a number: '))
    b = int(input('Enter a number: '))
    c = a / b
except ZeroDivisionError:
    print(f'Error: Number cannot be divisible by 0.')
except ValueError:
    print(f'Error: Input field only accept number integer values.')
else:
    print(round(c, 2))
finally:
    print('Calculation completed!!')
```

```python
file_name = input('Enter your file name: ')
extension = input('Enter file extension: ')
try:
    with open(f'{file_name}.{extension}', 'r') as file:
        content = file.read()
except FileNotFoundError:
    print(f'Error: {file_name}.txt file not found in directory.')
else:
    print(content)
finally:
    print('Program ended successfully!!!')
```

```python
class AgeExceptionError(Exception):
    pass
```

```python
def checkAge(age):
    if age >= 18:
        print('You are eligible for vote!')
    else:
        raise AgeExceptionError(f'Error: {age} age person is not eligible for vote!')
```

```python
try:
    age = int(input('Enter your age: '))
    checkAge(age)
except AgeExceptionError as e:
    print(e)
except ValueError:
    print(f'Error: Age value must be an integer.')
```
